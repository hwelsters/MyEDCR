```python
import pandas
```

```python
a = pandas.read_json("data/a.json")
a = a[a["epsilon"] != "ratio"]
a["epsilon"] = a["epsilon"].apply(lambda x : round(x, 2))
```

```python
# Draw bar chart graphing two columns
a.plot.line(x="epsilon", y=["precision_1", "recall_1"], color=["blue", "red"])

# Draw a line in the middle
import matplotlib.pyplot as plt
plt.axhline(y=0.220779, color='blue', linestyle='--', alpha=0.5)
plt.axhline(y=0.566667, color='red', linestyle='--', alpha=0.5)
plt.show()
```
